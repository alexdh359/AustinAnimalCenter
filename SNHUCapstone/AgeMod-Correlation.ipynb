```python
import numpy as np
import pandas as pd
recordsIn = pd.read_csv('Austin_Animal_Center_Intakes.csv', encoding='latin-1')
recordsOut = pd.read_csv('Austin_Animal_Center_Outcomes.csv', encoding='latin-1')
```

```python
recordsIn.head()
```

```python
simRec = recordsIn[['Intake Type', 'Intake Condition', 'Animal Type', 'Sex upon Intake', 'Age upon Intake']]
```

```python
simRec.head()
```

```python
new = simRec["Age upon Intake"].str.split(" ", n = 1, expand = True) 
```

```python
for i in range(len(new)):
    if new[1][i] == 'weeks':
        new[0][i] = float(new[0][i])/52.0
    elif new[1][i] == 'months':
        new[0][i] = float(new[0][i])/12.0
    else:
        new[0][i] = new[0][i]      
```

```python
new.head(30)
```

```python
simRec["Age"]= new[0] 
simRec.drop(columns =["Age upon Intake"], inplace = True) 
```

```python
simRec.head()
```

```python
new1 = simRec["Sex upon Intake"].str.split(" ", n = 1, expand = True) 
simRec["Neut/Spay"]= new1[0] 
simRec["Gender"]= new1[1] 
simRec.drop(columns =["Sex upon Intake"], inplace = True) 
```
