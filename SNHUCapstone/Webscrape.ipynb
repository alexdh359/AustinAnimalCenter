```python
import numpy as np
import pandas as pd
```

```python
intakesURL = 'https://data.austintexas.gov/api/views/wter-evkm/rows.csv?accessType=DOWNLOAD'
outcomeURL = 'https://data.austintexas.gov/api/views/9t4d-g238/rows.csv?accessType=DOWNLOAD'
```

```python
recordsIn = pd.read_csv(intakesURL, encoding='latin-1')
recordsOut = pd.read_csv(outcomeURL, encoding='latin-1')
```

```python
recordsIn.head()
```

```python
recordsIn.to_csv('Austin_Animal_Center_Intakes.csv')
recordsOut.to_csv('Austin_Animal_Center_Outcomes.csv')
```

```python
strayIn = recordsIn[recordsIn["Intake Type"] == 'Stray']
```

```python
strayIn['DateTime'] = pd.to_datetime(strayIn['DateTime'])
```

```python
strayIn.head()
```

```python
start_date = '2019-01-01 00:00:00'
end_date = '2019-03-01 00:00:00'
mask = (strayIn['DateTime'] > start_date) & (strayIn['DateTime'] <= end_date)
strayJtF = strayIn.loc[mask]
```

```python
start_date = '2018-01-01 00:00:00'
end_date = '2019-01-01 00:00:00'
mask = (strayIn['DateTime'] > start_date) & (strayIn['DateTime'] <= end_date)
stray2018 = strayIn.loc[mask]
stray2018.shape
```

```python
straydogcat2018 = stray2018[stray2018["Animal Type"].isin(['Dog', 'Cat'])]
```

```python
#Remove unknown or overly general locations
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Austin (TX)']
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Unknown in Travis (TX)']
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Unknown in Austin (TX)']
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Travis (TX)']
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Pflugerville (TX)']
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Outside Jurisdiction']
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Manor (TX)']
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Lago Vista (TX)']
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Jonestown (TX)']
straydogcat2018 = straydogcat2018[straydogcat2018["Found Location"] != 'Del Valle (TX)']


```
