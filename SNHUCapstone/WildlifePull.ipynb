```python
import numpy as np
import pandas as pd
recordsIn = pd.read_csv('Austin_Animal_Center_Intakes.csv', encoding='latin-1')
recordsOut = pd.read_csv('Austin_Animal_Center_Outcomes.csv', encoding='latin-1')
```

```python
#look at Stray
recordsStrayIn = recordsIn[recordsIn["Intake Type"] == 'Stray']
```

```python
import matplotlib.pyplot as plt
animaltype = recordsStrayIn['Animal Type'].value_counts()
animaltype.index.name = 'Animal'
animaltype.plot(kind='bar',rot=20, title='Strays Brought In',figsize=(14,5), colormap = 'Paired');
```

```python
import datetime
date = datetime.date.today()
animalplot = animaltype.plot(kind='bar',rot=20, title='Strays Brought In Before ' +str(date) ,figsize=(14,5), colormap = 'Paired');
```
