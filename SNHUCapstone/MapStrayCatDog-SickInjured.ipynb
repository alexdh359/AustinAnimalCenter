```python
import numpy as np
import pandas as pd
import geocoder
import folium
from geopy.geocoders import GoogleV3
from geopy import geocoders
recordsIn = pd.read_csv('MergedStrayCatDog_LatLong_2018.csv', encoding='latin-1')
#use https://www.kaggle.com/daveianhickey/how-to-folium-for-maps-heatmaps-time-analysis for reference
```

```python
coord = recordsIn
```

```python
#2000 records
Austin_COORDINATES = (coord["Latitude"][1], coord["Longitude"][1])
data = coord
# for speed purposes
MAX_RECORDS = recordsIn['Intake Condition'].shape[0]

 
m = folium.Map(location=Austin_COORDINATES, zoom_start=12)
 
# I can add marker one by one on the map
for i in range(0,MAX_RECORDS):
    if data.iloc[i]['Intake Condition'] == 'Sick':
       folium.Circle(
          location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
          #popup=data.iloc[i]['name'],
          radius=10,
          color='green',
          fill=True,
          fill_color='crimson'
       ).add_to(m)
    elif data.iloc[i]['Intake Condition'] == 'Injured': #injured
        folium.Circle(
          location=[data.iloc[i]['Latitude'], data.iloc[i]['Longitude']],
          #popup=data.iloc[i]['name'],
          radius=10,
          color='red',
          fill=True,
          fill_color='crimson'
       ).add_to(m)
  
display(m)
```

```python
m.save('mymap.html')
```
